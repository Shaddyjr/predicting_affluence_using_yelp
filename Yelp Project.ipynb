```python
from bs4 import BeautifulStoneSoup
import requests
import json
import pandas as pd
import numpy as np
import itertools
pd.set_option('display.max_colwidth', 3000)
pd.set_option('display.max_rows', 999)
```

```python
df = pd.read_json('./yelp_dataset/business.json', lines = True)
```

```python
df["price"] = df["attributes"].apply(lambda d: d.get("RestaurantsPriceRange2") if type(d) == dict else np.nan)
df['bit_coin']=  df["attributes"].apply(lambda d: d.get("BusinessAcceptsBitcoin") if type(d) == dict else np.nan)
df['business_parking'] = df["attributes"].apply(lambda d: d.get("BusinessParking") if type(d) == dict else np.nan)
df['appointment_only'] = df["attributes"].apply(lambda d: d.get("ByAppointmentOnly") if type(d) == dict else np.nan)
df['bike_parking'] = df["attributes"].apply(lambda d: d.get("BikeParking") if type(d) == dict else np.nan)
df['dogs_allowed'] = df["attributes"].apply(lambda d: d.get("DogsAllowed") if type(d) == dict else np.nan)
df['wheel_chair_access'] = df["attributes"].apply(lambda d: d.get("WheelchairAccessible") if type(d) == dict else np.nan)
df['restaurant_take_out'] = df["attributes"].apply(lambda d: d.get("RestaurantsTakeOut") if type(d) == dict else np.nan)
df['accept_credit_card'] = df["attributes"].apply(lambda d: d.get("BusinessAcceptsCreditCards") if type(d) == dict else np.nan)
df['noise_level'] = df["attributes"].apply(lambda d: d.get("NoiseLevel") if type(d) == dict else np.nan)
df['outdoor_seating'] = df["attributes"].apply(lambda d: d.get("OutdoorSeating") if type(d) == dict else np.nan)
df['kid_friendly'] = df["attributes"].apply(lambda d: d.get("GoodForKids") if type(d) == dict else np.nan)
df['smoking'] = df["attributes"].apply(lambda d: d.get("Smoking") if type(d) == dict else np.nan)
df['restaurant_attire'] = df["attributes"].apply(lambda d: d.get("RestaurantsAttire") if type(d) == dict else np.nan)
df['open_monday'] = df["hours"].apply(lambda d: d.get("Monday") if type(d) == dict else np.nan)
df['open_tuesday'] = df["hours"].apply(lambda d: d.get("Tuesday") if type(d) == dict else np.nan)
df['open_wednesday'] = df["hours"].apply(lambda d: d.get("Wednesday") if type(d) == dict else np.nan)
df['open_thursday'] = df["hours"].apply(lambda d: d.get("Thursday") if type(d) == dict else np.nan)
df['open_friday'] = df["hours"].apply(lambda d: d.get("Friday") if type(d) == dict else np.nan)
df['open_saturday'] = df["hours"].apply(lambda d: d.get("Saturday") if type(d) == dict else np.nan)
df['open_sunday'] = df["hours"].apply(lambda d: d.get("Sunday") if type(d) == dict else np.nan)
```

```python
days = ['open_monday', 'open_tuesday', 'open_wednesday', 'open_thursday', 'open_friday', 'open_saturday', 'open_sunday']
for i in days:
    df[i] = df[i].apply(lambda x: '0:0-0:0' if x == None else x)
    df[i] = df[i].apply(lambda x: 0 if x == '0:0-0:0' else 1)
```

```python
features = ['bike_parking', 
             'price', 
            'bit_coin', 
             'business_parking', 
             'appointment_only', 
             'dogs_allowed', 
            'wheel_chair_access', 
            'restaurant_take_out', 
            'noise_level', 
            'kid_friendly', 
            'smoking', 
            'restaurant_attire', 
            'accept_credit_card', 
            'outdoor_seating']
for i in features:
    df[i] = df[i].map({'True': 1, 'False': 0, 'None': 0})
df.fillna(0, inplace = True)
```

```python
df.drop(columns = ['hours', 'latitude', 'longitude'], inplace = True)
```
